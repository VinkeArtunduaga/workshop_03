```python
# Agregando la columna "year" a cada DataFrame
df_2015['year'] = 2015
df_2016['year'] = 2016
df_2017['year'] = 2017
df_2018['year'] = 2018
df_2019['year'] = 2019

# Concatenando los DataFrames
dataframes = [df_2015, df_2016, df_2017, df_2018, df_2019]
df_final = pd.concat(dataframes)
```

Aqui a cada uno de los df les agregue la columna llamada year para cada uno de los años respectivos.

```python
df = df_final.drop(["year"], axis=1)
df_numeric = df.select_dtypes(include=['number'])

correlation_matrix = df_numeric.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de correlación')
plt.show()
```

Aqui hago la visualizacion de la correlacion de las columnas en donde las que tuvieron amyor correlacion con el happiness_score fueron economy, health y social_support.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df_final[["economy", "social_support", "health"]]
y = df_final["happiness_score"]

# Divide los datos en conjuntos de training y testing (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=230)

model = LinearRegression()

model.fit(X_train, y_train)

# Calcula la precisión (R²) en el conjunto de testing
accuracy = model.score(X_test, y_test)

print("Precisión:", accuracy)
```

Aqui ya hago la division de los datos a partir del training y testing, este dispone de una precision del 70% aproximadamente
